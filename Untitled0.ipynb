## Data analysis

### Subtask:
Analyze the relationships between variables within each dataset and across datasets. Focus on crime data distribution.


**Reasoning**:
Analyze the distribution of crime types, crime distribution by location, temporal analysis, and relationships between datasets.  Calculate crime rates and identify potential issues with geographic data.



```python
# Crime Type Distribution
crime_type_counts = df_crimes['Primary Type'].value_counts()
print("Crime Type Distribution:\n", crime_type_counts)

# Crime Distribution by Location (Community Areas)
if 'Community Area' in df_crimes.columns and 'AREA_NUMBE' in df_comm_areas.columns:
    merged_comm = pd.merge(df_crimes, df_comm_areas, left_on='Community Area', right_on='AREA_NUMBE', how='left')
    crime_rates_comm = merged_comm.groupby('COMMUNITY')['ID'].count() / df_comm_areas.groupby('COMMUNITY')['AREA'].mean()
    print("\nCrime Rates per Community Area:\n", crime_rates_comm)
else:
    print("\nError: 'Community Area' or 'AREA_NUMBE' columns not found for community area analysis.")

# Crime Distribution by Location (Wards)
if 'Ward' in df_crimes.columns and 'WARD' in df_wards.columns:
    merged_wards = pd.merge(df_crimes, df_wards, left_on='Ward', right_on='WARD', how='left')
    crime_rates_wards = merged_wards.groupby('WARD')['ID'].count() / df_wards.groupby('WARD')['SHAPE_Area'].mean()
    print("\nCrime Rates per Ward:\n", crime_rates_wards)
else:
    print("\nError: 'Ward' or 'WARD' columns not found for ward analysis.")

# Crime Distribution by Location (Neighborhoods) - No direct identifier available, skipping

# Temporal Analysis
df_crimes['Date'] = pd.to_datetime(df_crimes['Date'])
daily_crimes = df_crimes.groupby(df_crimes['Date'].dt.date)['ID'].count()
weekly_crimes = df_crimes.groupby(df_crimes['Date'].dt.isocalendar().week)['ID'].count()
monthly_crimes = df_crimes.groupby(df_crimes['Date'].dt.month)['ID'].count()

print("\nDaily Crime Counts:\n", daily_crimes.head())
print("\nWeekly Crime Counts:\n", weekly_crimes.head())
print("\nMonthly Crime Counts:\n", monthly_crimes)

# Relationships between Datasets
# Example: Comparing crime rates with community area population (if available)
print("\nObservations on relationships between datasets:")
if 'POPULATION' in df_comm_areas.columns and 'Community Area' in df_crimes.columns:
    print("Further analysis could explore relationship between crime rates and community area population.")
else:
    print("Population data or 'Community Area' column is not found for correlational analysis.")
```

## Data visualization

### Subtask:
Visualize the key findings from the data exploration and analysis steps.


**Reasoning**:
Generate visualizations for crime type distribution, crime rates by location, and temporal trends, combining them into subplots for better presentation.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15))

# Crime Type Distribution
plt.subplot(2, 2, 1)
crime_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45, ha='right')

# Crime Rates by Community Area
plt.subplot(2, 2, 2)
crime_rates_comm.plot(kind='bar', color='lightcoral')
plt.title('Crime Rates per Community Area')
plt.xlabel('Community Area')
plt.ylabel('Crime Rate')
plt.xticks(rotation=45, ha='right')

# Crime Rates by Ward
plt.subplot(2, 2, 3)
crime_rates_wards.plot(kind='bar', color='lightgreen')
plt.title('Crime Rates per Ward')
plt.xlabel('Ward')
plt.ylabel('Crime Rate')
plt.xticks(rotation=45, ha='right')

# Temporal Trends (Daily, Weekly, Monthly)
plt.subplot(2, 2, 4)
plt.plot(daily_crimes.index, daily_crimes.values, label='Daily', color='blue')
plt.plot(weekly_crimes.index, weekly_crimes.values, label='Weekly', color='red')
plt.plot(monthly_crimes.index, monthly_crimes.values, label='Monthly', color='green')
plt.title('Temporal Trends of Crimes')
plt.xlabel('Time Period')
plt.ylabel('Number of Crimes')
plt.legend()

plt.tight_layout()
plt.show()
```
